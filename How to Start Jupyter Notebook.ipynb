## **Goals**

This tutorial is for macOS + Windows and we will go step by step on how to get Jupyter Notebook on your computer so we can start coding! This tutorial assumes that you _do not_ have anaconda downloaded or access to jupyter notebook. 



_What is Jupyter Notebook ?_

The [Jupyter Notebook](https://jupyter.org/) is the original web application for creating and sharing computational documents. It offers a simple, streamlined, document-centric experience. Jupyter supports over 40 programming languages, including Python, R, Julia, and Scala. In our tutorials we will be using motsly __Python__.

_Why use Jupyter Notebook for analyzing climate data ?_

Jupyter Notebook allows for interactive exploration and visualization of large climate datasets, seamlessly combining code, explanatory text, and visuals within a single document. Jupyter Notebook is the ideal place for iterative analysis, data cleaning, and communicating climate research findings clearly to both technical and non-technical audiences.

## **Download Anaconda**

To Start we need to use Anaconda to accesss Jupyter Notebook by downloading the anaconda in either your macOS terminal or windows command prompt window. 

_What is a mac terminal ?_

The Mac Terminal is a command-line interface (CLI) that allows users to interact with the Mac operating system. It is an application found on your macOS.



_What is the Window's Command Prompt Window?_

Command Prompt" or "CMD," is a text-based interface on a Windows computer where users can type commands to interact with the operating system and perform various tasks. It is an application found on your Windows PC



_What is Anaconda ?_

[Anaconda](https://www.anaconda.com/) is an open-source distribution of the Python programming language, specifically designed for data science and machine learning, which provides a collection of pre-installed packages and a package manager called __"conda"__ to simplify the installation and management of various data science libraries we will use in our climate data lab.
